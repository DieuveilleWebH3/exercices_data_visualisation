**Exercice 2:** Visualisation de Relations Continues 



```python
import pandas as pd
import numpy as np
import seaborn as sns
from seaborn import relplot
from sklearn import linear_model
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
# read sales_predictions.csv data in pandas dataframe df_sales using pandas read_csv function
df_sales = pd.read_csv('C:/Users/USER/Desktop/vm_test/H3Hitema/Analyse de donnee Python/Datasets/sales_predictions.csv')
```

```python
df_sales.head()
```

```python
# sns.relplot(data=df_sales, x="date", y="item_price")

# sns.relplot(data=df_sales.sample(50), x="date", y="item_price")

sns.relplot(data=df_sales.sample(50), x="date", y="item_price", hue='item_id', sizes=(40, 400))

plt.show()
```

```python
sns.set_theme(style="white")
sns.set(style="white")
sns.set_palette('viridis')
sns.set(rc={'figure.figsize':(20, 6)}) 
```

```python
sns.relplot(data=df_sales.sample(50), x="date", y="item_price")

plt.show()
```

```python
# from pandas import datetime
# from datetime import datetime
from pandas import to_datetime
```

```python
df_date = to_datetime(df_sales.date) 
# sns.relplot(data=df_sales.sample(50), x=df_date, y="item_price")

sns.relplot(data=df_sales.sample(50), x=df_date, y="item_price", sizes=(40, 400), alpha=.5, palette="muted", height=6)

plt.show()
```
