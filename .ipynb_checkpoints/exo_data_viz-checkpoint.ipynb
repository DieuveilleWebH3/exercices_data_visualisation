**Exercice 2:** Visualisation de Relations Continues 



```python
import pandas as pd
import numpy as np
import seaborn as sns
from seaborn import relplot
from sklearn import linear_model
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.cbook as cbook
import csv
%matplotlib inline
```

```python
# read sales_predictions.csv data in pandas dataframe df_sales using pandas read_csv function
df_sales = pd.read_csv('C:/Users/USER/Desktop/vm_test/H3Hitema/Analyse de donnee Python/Datasets/sales_predictions.csv')
```

```python
df_sales.head()
```

```python
# sns.relplot(data=df_sales, x="date", y="item_price")

# sns.relplot(data=df_sales.sample(50), x="date", y="item_price")


sns.relplot(data=df_sales.sample(50), x="date", y="item_price", hue='item_id', height=6.7, sizes=(40, 400))

fig=plt.figure(figsize=(20,6), dpi= 100)
plt.rcParams["figure.figsize"] = (20, 6)

plt.show()

```

```python
file = 'C:/Users/USER/Desktop/vm_test/H3Hitema/Analyse de donnee Python/Datasets/sales_predictions.csv'

x = []
y = []
  
i = 0

with open(file,'r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
      
    for row in plots:
        if i >= 50:
            break
            
        x.append(row[0])
        y.append(row[4])
        
        i = i + 1

x = sorted(x)
y = sorted(y)

plt.bar(x, y, color = 'g', width = 0.72, label = "date")
plt.xlabel('date')
plt.ylabel('prix')
plt.title('Prix par Date')
plt.legend()
plt.show()
```

```python
sns.set_theme(style="white")
sns.set(style="white")
sns.set_palette('viridis')
sns.set(rc={'figure.figsize':(20, 6)}) 
```

```python
sns.relplot(data=df_sales.sample(50), x="date", y="item_price", height=8.7)

plt.show()
```

```python
# from pandas import datetime
# from datetime import datetime
from pandas import to_datetime
```

```python
df_date = to_datetime(df_sales.date) 
# sns.relplot(data=df_sales.sample(50), x=df_date, y="item_price")

sns.relplot(data=df_sales.sample(50), x=df_date, y="item_price", sizes=(40, 400), alpha=.5, palette="muted", height=8.7)

plt.show()
```
